```python
from pyspark import SparkContext
from pyspark.streaming import StreamingContext
import pyspark as ps

sc = ps.SparkContext('local[*]')
ssc = StreamingContext(sc, 5)
```

```python
file = open('../data/streamRDDs.txt', 'r').read()
sentences = file.split('.\n')
rddQueue = []
for sentence in sentences:
    rddQueue.append(sc.parallelize(sentence.split(' ')))

inputStream = ssc.queueStream(rddQueue)
```

```python
pairs = inputStream.map(lambda word: (word, 1))
wordCounts = pairs.reduceByKey(lambda num1, num2: num1 + num2)
# find the word with the maximum count
maxCount = wordCounts.reduce(lambda x, y: x if x[1] > y[1] else y)
# print the word with the maximum count
maxCount.pprint()
```

```python
# find the most long word
longestWord = wordCounts.reduce(lambda x, y: x if len(x[0]) > len(y[0]) else y)
# print the most long word
longestWord.pprint()
```

```python
ssc.start()
ssc.awaitTermination()
```
